# Linear Regression using a Single Features

```python
non_smoker_df = medical_df[medical_df.smoker == 'no']
non_smoker_df.shape
```

```python
plt.title('Age Vs. Charges for Non-Smokers')
sns.scatterplot(data = non_smoker_df, x = 'age', y = 'charges', alpha=0.7, s = 15)
```

## Model 

```python
def estimate_charges(age, w, b):
    return w*age + b
```

```python
w = 50
b = 100
```

```python
ages = non_smoker_df.age
estimated_charges = estimate_charges(ages, w, b)
```

```python
plt.plot(ages, estimated_charges, 'r-o')  
plt.xlabel('Age')
plt.ylabel('Charges')
```

```python
target = non_smoker_df.charges

plt.plot(ages, estimated_charges, 'r', alpha=0.9);
plt.scatter(ages, target, s=8,alpha=0.8);
plt.xlabel('Age');
plt.ylabel('Charges')
plt.legend(['Estimate', 'Actual']);
```

```python
target = non_smoker_df.charges

plt.plot(ages, estimated_charges, 'r', alpha=0.9);
plt.scatter(ages, target, s=8,alpha=0.8);
plt.xlabel('Age');
plt.ylabel('Charges')
plt.legend(['Estimate', 'Actual']);
```

```python
def try_parameters(w, b):
    ages = non_smoker_df.age
    target = non_smoker_df.charges
    
    estimated_charges = estimate_charges(ages, w, b)
    
    plt.plot(ages, estimated_charges, 'r', alpha=0.9);
    plt.scatter(ages, target, s=8,alpha=0.8);
    plt.xlabel('Age');
    plt.ylabel('Charges')
    plt.legend(['Estimate', 'Actual']);
```

```python
try_parameters(60, 200)
```

```python
try_parameters(400, 5000)
```

## Using Train-Test Split Method

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
inputs = medical_df[['age']]
targets = medical_df['bmi']
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs, targets, test_size=0.2, random_state=42) # 80% training data, 20% testing data
print('X_train shape:', X_train.shape)
print('X_test shape:', X_test.shape)
print('y_train shape:', y_train.shape)
print('y_test shape:', y_test.shape)
```

## Create a model

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

# Evaluation of the Module 

```python
mse = mean_squared_error(y_test, y_pred)
print(mse)
```

```python
r2_score = r2_score(y_test, y_pred)
print(r2_score)
```
